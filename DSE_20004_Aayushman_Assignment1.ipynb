# Scraping ICML workshop website

```python
# Load the modules
from gazpacho import get, Soup
import numpy as np
import pandas as pd
```

```python
# Load the website
url = "https://icml.cc/virtual/2022/events/workshop"
```

```python
# Soupify the url
soup = Soup(get(url))
```

`Now We try to obtain different type of entity from the website`

Namely: Title of the workshop, author and the abstract

```python
abstract = [i.text for i in soup.find('p')]
```

```python
a = [j.text for j in soup.find('div',{'class':'author-str'})]
author = []
for i in a:
    if len(i)>0:
        author.append(i)
```

```python
title = [k.text for k in soup.find('a',{'class':'small-title'})]
```

```python
list_of_author=[]
for i in range(0,33):
    list_of_author.append(author[i].split('·'))
    
```

```python
list_of_author
```

```python
dataframe1 = pd.DataFrame(title,columns=['Title'],index=range(1,35))
dataframe2 = pd.DataFrame(author,columns=['Author'],index=range(1,35))
```

```python
Data = pd.concat([dataframe1,dataframe2],axis=1)
```
